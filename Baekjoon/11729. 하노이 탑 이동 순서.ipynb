```python
import sys

sys.stdin = open('11729. 하노이 탑 이동 순서.txt', 'rt')

def hanoi_tower(n, start, end) :
    if n == 1 :
        print(start, end)
        return
       
    hanoi_tower(n-1, start, 6-start-end) # 1단계
    print(start, end) # 2단계
    hanoi_tower(n-1, 6-start-end, end) # 3단계
    
n = int(sys.stdin.readline())
print(2**n-1)
hanoi_tower(n, 1, 3)
```
